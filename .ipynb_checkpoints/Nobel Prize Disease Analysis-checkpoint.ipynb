# Nobel Prize Disease 

##### Purpose
In this notebook we explore some basic questions about the so-called Nobel Prize Disease.
This points to those Nobel laureates that supported some sort of pseudoscientific crankery, and the term is coined precisely because
one would think that these high level scientists would be the most immune to such beliefs.  The goal of this notebook,
however, is to actually point out the contrary, i.e. that the population of Nobel laureates contains, in fact, a remarkably
low number of individuals with such beliefs when we compare them to the general population.

# First exploration of the Nobel dataset <a id=intro><a>

### Import the data <a id=import><a>

```python
df = pd.read_csv("dataset.csv")
```

```python
df.head()
```

```python
df_disease = df[df['Disease']=='yes']
df_disease
```

### By Country and Category <a id=basicstats><a>

```python
countries = df_disease.groupby('Birth Country').size()
countries.plot.pie(figsize=(4,4), title="Nobel Diseases by Country")
```

```python
discipline = df_disease.groupby('Category').size()
discipline.plot.pie(figsize=(4,4), title="Nobel Diseases by Category")
```

```python
decade_df = df_disease
decade_df['decade'] = round(decade_df['Year']/10)*10
decade = decade_df.groupby('decade').size()
decade.plot.bar(figsize=(4,4), title="Nobel Diseases by Decade")
```

### Prevalence <a id=prevalence><a>

The percentage of laureates that believe in some nonsense:

```python
percentage = round(df_disease.count()[1]/df.count()[1]*10000)/100
print(str(percentage) + " %")
```

Paranormal:

```python
count_paranormal = sum(df_disease['Diagnostic'].apply(lambda x: 'aranormal' in x))
percentage_paranormal = round(count_paranormal/df.count()[1]*10000)/100
print( 'Absolute count: ' + str(count_paranormal))
print('Percentage: ' + str(percentage_paranormal) + " %")
```

Homeopathy:

```python
count_homeopathy = sum(df_disease['Diagnostic'].apply(lambda x: 'omeopathy' in x))
percentage_homeopathy = round(count_homeopathy/df.count()[1]*10000)/100
print( 'Absolute count: ' + str(count_homeopathy))
print('Percentage: ' + str(percentage_homeopathy) + " %")
```

# Research on general population <a id=research><a>

It is hard to find any research on general pseudoscience beliefs in the overall population, but I did manage to find a few interesting articles for paranormal and homeopathy categories.

Paranormal:

There is a fascinating paper published in Heliyon in 2017 by Bernard Garrett and Roger Cutting named "Magical beliefs and discriminating science from pseudoscience in undergraduate professional students".  This is a fascinating read which studies the relationship of different kinds of media exposure and subsequent pseudoscientific beliefs.
One can see here that belief in paranormal varies across countries in the figures they cite, but around 40% of population believing in ghosts/paranormal seems reasonable.

Homeopathy:

Belief in homeopathy varies greatly by country. In Germany it's 11.5% according to 

Bücker, B; Groenewold, M; Schoefer, Y; Schäfer, T (2008). "The use of complementary alternative medicine (CAM) in 1001 German adults: Results of a population-based telephone survey".

In the US, a 2012 Survey estimated the prevalence at around 2.1%:

"Homeopathy Use by US Adults: Results of a National Survey" by Michelle L. Dossett, MD, PhD, MPH,corresponding author Roger B. Davis, ScD, Ted J. Kaptchuk, and Gloria Y. Yeh, MD, MPH

# Conclusion <a id=conclusion><a>

We see that in the case of homeopathy and the paranormal beliefs, the Nobel Prize population is at least an order of magnitude less susceptible of falling for the pseudosciences than the general population.  Although it is hard to understand how a highly educated and brilliant individual may fall for these ideas, one must consider that 

(a) Many of these scientists are from early 20th century when information was not so ready accessible.

(b) A few of these expressed their ideas in their senile age (e.g. Luc Montagnier saying nonsense at age 77) or were influenced by some heavy drug addictions (Kary Mullis was a heavy LSD user).

```python
prevalences = {"Population":["Homeopathy Nobel", "Homeopathy General", "Paranormal Nobel", "Paranormal General"],

        "Prevalence":[0.31, 2.1, 0.93, 40]

        };

# Dictionary loaded into a DataFrame       

prevalences_df = pd.DataFrame(data=prevalences);
```

```python
prevalences_df.plot.bar(x="Population", y="Prevalence", rot=100, title="Prevalence of beliefs by population");
```

All the above figures are not meant to be exact measurements.  In the chart above I used the 2.1% prevalence of homeopathy in the US (a lower number than most european countries).  But the ball park figures already give us an idea of how drastically different the beliefs are between these two groups.
